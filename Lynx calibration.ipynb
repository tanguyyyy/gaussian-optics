```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def extraction(nom_fichier):
    L=[]
    list=[]
    f=open(nom_fichier,"r")
    for ligne in f:
        L.append(ligne.split(" "))
    for i in range (len(L)):
        if i==0 :
            for j in range (len(L[0])-2):
                list.append(int(L[i][j]))
        if i>0 and i<len(L)-1 :
            list.append(int(L[i][0].replace('\t\t','')))
            for j in range (1,len(L[0])-1):
                list.append(int(L[i][j]))
        else :
            list.append(int(L[i][0].replace('\t\t','')))
            for j in range (1,len(L[i])-1):
                list.append(int(L[i][j]))
    return list
```

```python
spectrum = extraction('cobalt57_1_manual.txt')
```

```python
plt.subplots(figsize=(15, 10))

plt.subplot(2,2,1)
plt.plot(np.arange(1,len(spectrum)+1),spectrum, 'green')
plt.xlabel('Channels')
plt.ylabel('Counts (log)')
plt.yscale('log')
plt.title('Cobalt 57 (log) entier')

plt.subplot(2,2,2)
plt.plot(np.arange(1,len(spectrum)+1),spectrum)
plt.xlabel('Channels')
plt.ylabel('Counts')
plt.title('Cobalt 57 entier')

plt.subplot(2,2,3)
plt.plot(np.arange(1,len(spectrum)+1),spectrum,'green')
plt.xlim(2700,3100)
plt.xlabel('Channels')
plt.ylabel('Counts (log)')
plt.yscale('log')
plt.title('Cobalt 57 (log) zoomé')

plt.subplot(2,2,4)
plt.plot(np.arange(1,len(spectrum)+1),spectrum)
plt.xlim(2700,3100)
plt.ylim(0,800)
plt.xlabel('Channels')
plt.ylabel('Counts')
plt.title('Cobalt 57 zoomé')

plt.tight_layout()

plt.show()
```

**Tentative de fit**

```python
from scipy.optimize import curve_fit
```

```python
def func(x, a, m, s):

    return a * np.exp(-((x - m)**2)/(2*(s**2)))
```

```python
spectrum = extraction('cobalt57_1_manual.txt')
```

```python
peak1 = spectrum[1400:1450]
```

```python
plt.plot(peak1)
```

```python
popt, pcov = curve_fit(func, np.arange(1400,1450), np.array(peak1), p0=np.array([500000,25,10]), bounds=(-np.inf, np.inf))

popt
```

```python

plt.plot(np.arange(1400,1450), func(np.arange(1400,1450), *popt))
```

```python
def Gauss(x, A, B, C):
    y = A*np.exp(-1*B*x**2)+C
    return y
```

```python
#peak1
parameters, covariance = curve_fit(Gauss, np.arange(1300,1500), peak1)
fit_A = parameters[0]
fit_B = parameters[1]
fit_C = parameters[2]
print(fit_A, fit_B, fit_C)
  
fit_y = Gauss(np.arange(1300,1500), fit_A, fit_B, fit_C)
plt.plot(np.arange(1300,1500), peak1, 'o', label='data')
plt.plot(np.arange(1300,1500), fit_y, '-', label='fit')
plt.legend()
```

```python
from scipy import asarray as ar,exp
x = np.arange(1300,1500)
y = np.array(peak1)
n = len(x)                          #the number of data
mean = sum(x*y)/n                   
sigma = sum(y*(x-mean)**2)/n        

def gaus(x,a,x0,sigma):
    return a*exp(-(x-x0)**2/(2*sigma**2))

popt,pcov = curve_fit(gaus,x,y,p0=[1,mean,sigma])

plt.plot(x,y,'b+:',label='data')
plt.plot(x,gaus(x,*popt),'ro:',label='fit')
plt.legend()
plt.title('Fig. 3 - Fit for Time Constant')
plt.xlabel('Channels')
plt.ylabel('Counts')
plt.show()
```
